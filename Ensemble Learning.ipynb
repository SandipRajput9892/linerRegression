```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=500,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
dataset=pd.DataFrame(df)
```

```python
sns.scatterplot(x="x1",y="x2",data=dataset,hue="y")
plt.show()
```

```python
x_a=dataset.iloc[:,:-1]
y_a=dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100,dt.score(x_test,y_test)*100
# dt.score(x_test,y_test)*100
# dt.predict([])
```

```python
sv=SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_train,y_train)*100,gnb.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li=[("dt",DecisionTreeClassifier()),("sv",SVC()),("gnb",GaussianNB())]
```

```python
vc=VotingClassifier(li,weights=[3,4,3])
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100,vc.score(x_test,y_test)
```

```python
from sklearn.svm import SVC

sv = SVC()  # You can specify kernel or other parameters here
sv.fit(x_train, y_train)
```

```python
prd={"dt":dt.predict(x_test),"sv":sv.predict(x_test),"gnb":gnb.predict(x_test),"vc":vc.predict(x_test)}
prd
```
