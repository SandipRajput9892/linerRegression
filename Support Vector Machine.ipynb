```python
import pandas as pd
from sklearn.model_selection import train_test_split
# from sklearn.metrics import root_mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("placement (1).csv")
df=pd.DataFrame(dataset)
df.head(5)
```

```python
sns.scatterplot(x="cgpa",y="resume_score",data=dataset,hue="placed")
plt.show()
```

```python

x=df.drop(["placed"],axis=1)
y=df["placed"]
# x_train,x_test,y_tarin,y_test=train_test_split(x,y,train_size=0.2)
```

```python
x_train,x_test,y_tarin,y_test=train_test_split(x,y,train_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_tarin)
```

```python
model.score(x_test,y_test)*100
```

```python
model.predict([[7.52,7.30]])
```
